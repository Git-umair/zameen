```python
import requests
from bs4 import BeautifulSoup

url = 'https://cjdropshipping.com/list/wholesale-baby-rompers-l-80304DEB-99FB-4E29-9065-A99F732702C4.html?from=US'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

inner_boxes = soup.find_all('div', class_='inner-box')
for box in inner_boxes:
    print(box.text.strip())
    links = box.find_all('a')
    for link in links:
        print(link.get('href'))
```

```python
import requests
from bs4 import BeautifulSoup

url = 'https://cjdropshipping.com/list/wholesale-baby-rompers-l-80304DEB-99FB-4E29-9065-A99F732702C4.html?from=US'

response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

inner_boxes = soup.find_all('div', class_='inner-box')
for box in inner_boxes:
    # extract the text content
    text = box.text.strip()
    print(text)

    # extract the link URL
    link = box.find('a')['href']
    print(link)

    # add a separator between entries for readability
    print('-' * 40)
```
